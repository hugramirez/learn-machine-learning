```python
import pandas as pd
import numpy as np
from typing import Dict, Any
```

```python
dtypes = {
    "gender": "category",
    "age_range": "category"
}

parse_dates = ["start_dt", "end_dt", "start_date", "end_date"]

dataframe = pd.read_csv(
    "/Users/psf/Github/learn-machine-learning/data/raw/sistema_ecobici_cdmx.csv",
    dtype=dtypes, # type: ignore
    parse_dates=parse_dates,
    index_col=False,
    engine='python',
    chunksize=100000
)
```

```python
sample = pd.concat(dataframe, ignore_index=True)
```

```python
sample
```

```python
sample.info()
```

```python
if sample["age"].isna().sum() == 0:
    sample["age"] = sample["age"].astype("Int8")

sample["bike_id"] = sample["bike_id"].astype("Int32")
sample["start_hour"] = sample["start_hour"].astype("Int8")
sample["day"] = sample["day"].astype("Int8")
sample["age"] = sample["age"].astype("Int8")
```

```python
sample.isnull().sum()
```

```python
sample[sample["age_range"].isnull()]
```

```python
sample.dropna(subset=["age_range"], inplace=True)
bins = [10, 14, 19, 24, 29, 34, 39, 44, 49, 54, 59, 64, 69, 74, 79, 84, 89, 94, 99, 104]
labels = [
    "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40-44",
    "45-49", "50-54", "55-59", "60-64", "65-69", "70-74", "75-79",
    "80-84", "85-89", "90-94", "95-99", "100-104"
]

sample['age_range'] = pd.cut(sample['age'], bins=bins, labels=labels, right=True)
sample['gender'] = sample['gender'].map({'hombres': 'male', 'mujeres': 'female', 'otros': 'other', '?': 'other'}).fillna('other')
```
